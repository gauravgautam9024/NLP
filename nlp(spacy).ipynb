```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Gaurav lives in India"
doc = nlp(text)

for token in doc:
    print(token.text, token.pos_)
```

# 1. Named Entity Recognition (NER):-
                                   important cheeze detect karna (person, country, company)

```python
import spacy
nlp = spacy.load("en_core_web_sm")
text = "Elon Musk founded Tesla in the USA"
doc = nlp(text)
for ent in doc.ents:
    print(ent.text, ent.label_)
```

# 2. Stop Words remove karna:-
                            Common words hata dete hain (is, the, in)

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Gaurav is working in a tech company"
doc = nlp(text)

for token in doc:
    if not token.is_stop:
        print(token.text)
```

# 3. Lemmatization (word ka base form):-
                                     words ko root form me convert karta hai

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "running played better"
doc = nlp(text)

for token in doc:
    print(token.text, "->", token.lemma_)
```

# 4. Sentence detection:-
                       paragraph ko sentences me todna

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Gaurav lives in India. He works in AI. He is the best"
doc = nlp(text)

for sent in doc.sents:
    print(sent)
```

# 5. Dependency Parsing (kaun kya kar raha hai):-
                                                subject, object samajhna

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Gaurav eats mango"
doc = nlp(text)

for token in doc:
    print(token.text, token.dep_)
```

# 🔹 6. Noun chunks (group of words):- 
                                        noun phrases nikalna

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Gaurav bought a new laptop"
doc = nlp(text)

for chunk in doc.noun_chunks:
    print(chunk)
```

# 7. Custom filtering (sirf nouns nikaalo)

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Gaurav is learning machine learning"
doc = nlp(text)

for token in doc:
    if token.pos_ == "NOUN":
        print(token.text)
```
