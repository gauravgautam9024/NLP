```python
import pandas as pd
import numpy as np
import nltk
import re
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
nltk.download('stopwords')
```

```python
# data = {
#     "text": [
#         "I am very happy today",
#         "This is amazing",
#         "I feel great",
#         "I am so excited",
#         "I am sad",
#         "This is depressing",
#         "I feel bad",
#         "I am unhappy",
#         "I am angry",
#         "This makes me mad",
#         "I am frustrated",
#         "I hate this"
#     ],
#     "emotion": [
#         "happy", "happy", "happy", "happy",
#         "sad", "sad", "sad", "sad",
#         "angry", "angry", "angry", "angry"
#     ]
# }

data = {
    "text": [
        # 😊 Happy
        "I am feeling very happy today",
        "This is such a wonderful day",
        "I feel joyful and excited",
        "Life is beautiful",
        
        # 😢 Sad
        "I feel very sad",
        "This is depressing",
        "I am heartbroken",
        "I feel lonely and down",
        
        # 😡 Angry
        "I am very angry right now",
        "This makes me mad",
        "I am frustrated",
        "I hate this situation",
        
        # 😨 Fear
        "I am scared of this",
        "This is terrifying",
        "I feel nervous",
        "I am afraid of what will happen",
        
        # 😲 Surprise
        "Wow this is surprising",
        "I did not expect this",
        "This shocked me",
        "I am amazed",
        
        # 🤢 Disgust
        "This is disgusting",
        "I feel sick seeing this",
        "This is gross",
        "I hate this smell",
        
        # 😍 Love
        "I love you so much",
        "This is lovely",
        "I feel so attached",
        "You mean everything to me",
        
        # 😌 Calm
        "I feel relaxed",
        "This is peaceful",
        "I am calm and quiet",
        "I feel at ease"
    ],
    
    "emotion": [
        "happy","happy","happy","happy",
        "sad","sad","sad","sad",
        "angry","angry","angry","angry",
        "fear","fear","fear","fear",
        "surprise","surprise","surprise","surprise",
        "disgust","disgust","disgust","disgust",
        "love","love","love","love",
        "calm","calm","calm","calm"
    ]
}
df = pd.DataFrame(data)
df
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]',' ', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

df['clean_text'] = df['text'].apply(clean_text)
df
```

```python
# Step 4: Tokenization & Padding

tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['clean_text'])    #number mapping for every word

X = tokenizer.texts_to_sequences(df['clean_text'])      #Ye line text ko numbers me convert karti hai
X = pad_sequences(X, maxlen=10)                            #Ye sab sequences ko same length ka banata hai maxlen = 10 => [0,0,0,0,0,0,0,2,3,1]
```

```python
# Step 5: Label Encoding

encoder = LabelEncoder()
y = encoder.fit_transform(df['emotion'])
```

```python

# Step 6: Build Model (Deep Learning)

model = Sequential()           # Simple stack model (layer by layer)
model.add(Embedding(input_dim=1000, output_dim=64, input_length=10))       #Ye words ko numbers → vectors me convert karta hai
model.add(LSTM(64))       # Ye sequence (sentence) ko samajhta hai (context) Example: "not good" vs "good" → LSTM difference samajh lega
model.add(Dense(32, activation='relu'))
model.add(Dense(8, activation='softmax'))   # ab multiple classes (3 ya usse zyada) ho Har class ki probability deta hai . Example output:[0.1, 0.7, 0.2]

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# Step 7: Train Model

model.fit(X, y, epochs=20, batch_size=4, verbose=0)
```

```python
# Step 8: Emotion Prediction Function

def predict_emotion(text):
    text = clean_text(text)
    seq = tokenizer.texts_to_sequences([text])
    padded = pad_sequences(seq, maxlen=10)
    
    pred = model.predict(padded)
    emotion = encoder.inverse_transform([np.argmax(pred)])[0]
    
    return emotion
```

```python
# Step 9: Take User Input

while True:
    user_input = input("\nEnter your sentence (type 'exit' to quit): ")
    
    if user_input.lower() == 'exit':
        print("Exiting Emotion Detector ")
        break
    
    emotion = predict_emotion(user_input)
    print(f"Detected Emotion: {emotion}")
```
