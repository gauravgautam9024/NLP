# 1️⃣ Text Cleaning
Text cleaning ka matlab hai text se unwanted elements remove karna.

Unwanted elements ho sakte hain:

Capital letters

Punctuation

Numbers

Special characters

Emojis

Extra spaces

Example raw text:
"Hello!!! This Phone Cost 20,000 Rupees!!! 😍"
Clean text:
hello this phone cost rupees
Subtopics of Text Cleaning
1. Lowercasing
Lowercasing ka matlab hai sabhi letters ko small letters me convert karna.

Example:
I Love Data Science

Lowercase:
i love data science

Reason:

Machine learning models "Apple" aur "apple" ko different words samajh sakte hain, jo problem create karta hai.

Example problem:

Apple apple APPLE1

# 2️⃣ Removing Punctuation
Punctuation marks language ke liye useful hote hain, lekin machine learning models ke liye usually unnecessary hote hain.

Examples of punctuation:
1
. , ! ? : ; ' "

Example sentence:

Hello!!! How are you???

After removing punctuation:

Hello How are you1

```python
import string

text = "Hello!!! How are you???"

for p in string.punctuation:
    text = text.replace(p,"")

print(text)
```

# 3️⃣ Removing Numbers
Kabhi kabhi text me numbers important hote hain, lekin many NLP tasks me numbers noise create karte hain.

Example:

This phone costs 20000 rupees

After removing numbers:

This phone costs rupees1

```python
import re

text = "This phone costs 20000 rupees"

clean = re.sub(r'\d+', '', text)

print(clean)
```

```python
4️⃣ Removing Special Characters
Special characters bhi text ko noisy banate hain.

Examples:

@ # $ % ^ & * ( ) _

Example text:

I love #DataScience!!!

Clean text:

I love DataScience1
```

```python
import re

text = "I love #DataScience!!!"

clean = re.sub(r'[^a-zA-Z\s]', '', text)

print(clean)
```

```python
2️⃣ Tokenization
Tokenization NLP ka fundamental step hai.

Definition:

Sentence ko small units (tokens) me break karna

Tokens ho sakte hain:

Words Sentences Characters

Most common: Word Tokenization1
```

```python
from nltk.tokenize import word_tokenize

text = "I love data science"

tokens = word_tokenize(text)

print(tokens)
```

# 3️⃣ Stopwords Removal
Stopwords wo words hote hain jo language me bahut common hote hain lekin meaning me zyada contribution nahi dete.

Examples

is the and a in of to

Example sentence:

This is the best phone in the market

After removing stopwords:

best phone market

Isse important words hi bach jate hain.

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

words = ["this","is","a","good","phone"]

filtered = []

for w in words:
    if w not in stop_words:
        filtered.append(w)

print(filtered)
```

# 4️⃣ Stemming
Stemming me word ko root form tak cut kar diya jata hai.

Ye rule-based hota hai aur kabhi kabhi incorrect words bhi bana deta hai.

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

print(ps.stem("playing"))
print(ps.stem("running"))
print(ps.stem("studies"))
```

# 5️⃣ Lemmatization
Lemmatization stemming ka advanced version hai.

Ye dictionary aur grammar rules use karta hai taaki correct root word mile.
pos tags :
'n' Noun(Naam)
'v' Verb(kaam)
'a' Adjeactive(sifat)
'r' Adverb(kaam ki quality)

```python
from nltk.stem import WordNetLemmatizer

lemm = WordNetLemmatizer()

print(lemm.lemmatize("running",pos="v"))
print(lemm.lemmatize("darker",pos="a"))
```
