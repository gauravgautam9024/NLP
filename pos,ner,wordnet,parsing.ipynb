# Part of Speech (POS) Tagging

POS Tagging ka matlab hota hai sentence ke har word ko uski grammatical category assign karna.

Simple language me: 👉 Har word ko batana ki wo noun hai, verb hai, adjective hai, etc.
Example :this is gaurav
[('this', 'DT'), ('is', 'VBZ'), ('gaurav', 'JJ')]

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize

text = input("Word input - ")
tokens = word_tokenize(text)

print(pos_tag(tokens))
```

# Named Entity Recognition (NER)
Named Entity Recognition (NER) NLP ka ek important task hai jisme hum text me se important entities (names) identify karte hain.

Simple words me:

NER ka kaam hota hai text me se “kaun, kaha, kya” nikalna.

Example:
Sentence:

"Elon Musk is the CEO of Tesla and lives in the United States."

👉 NER Output:

Elon Musk → Person

Tesla → Organization

United States → Location

Entity Type	         Meaning	                           Example
PERSON	            Insaan ka naam	                             Elon Musk
ORG	                Company/Organization	                     Tesla
GPE           	    Country/City	                             USA, India
DATE	            Date	                                     24 March
TIME	           Time	                                         10 AM
MONEY	           Paisa	                                     $100
PERCENT	           Percentage	                                 50%

# N-Gram
words ke groups (sequence) banana

                                                                    Type	      Meaning	   Example
                                                                    Unigram	       1 word	     I, love, NLP
                                                                    Bigram	       2 words	     I love, love NLP
                                                                    Trigram	        3 words	     I love NLP
Example

Sentence:

"I love NLP"

Output:

Unigram → I, love, NLP Bigram → (I love), (love NLP) Trigram → (I love NLP)

```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

text =input("Enter the input -  ")
tokens = word_tokenize(text)

# Bigrams
bigrams = list(ngrams(tokens, 2))
print(bigrams)
```

# WordNet (Synonyms & Antonyms)
WordNet NLTK ka ek lexical database (dictionary) hai jisme words ke:

Synonyms (same meaning)

Antonyms (opposite meaning)

Definitions

Relationships

store hote hain

Word: good

Synonyms → good, excellent, nice Antonyms → bad

```python
import nltk
nltk.download('wordnet')
```

```python
#Synonyms

from nltk.corpus import wordnet

synsets = wordnet.synsets("excellent")

for syn in synsets:
    for lemma in syn.lemmas():
        print(lemma.name())
```

```python
#Antonyms (Opposite Meaning)
from nltk.corpus import wordnet

word = "good"
synsets = wordnet.synsets(word)

antonyms = []

for syn in synsets:
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

print(set(antonyms))
```

```python
# meaning 
from nltk.corpus import wordnet

syn = wordnet.synsets("work")[0]
print(syn.definition())
```

# Parsing (Grammar Structure)
Parsing NLP ka ek process hai jisme hum sentence ka grammar structure (structure/relationship) samajhte hain.

Simple words me:

Parsing batata hai sentence me kaun subject hai, kaun verb hai, kaun object hai

Example Sentence

"I eat apple"

Parsing structure:

I → Subject (NP – Noun Phrase)

eat → Verb (VP – Verb Phrase)

apple → Object

Term	Meaning
S	     Sentence
NP	      Noun Phrase
VP	       Verb Phrase
V	          Verb


```python
import nltk

grammar = nltk.CFG.fromstring("""
S -> NP VP
NP -> 'I' | 'apple'
VP -> V NP
V -> 'eat'
""")
```

```python
parser = nltk.ChartParser(grammar)

sentence = ["I", "eat", "apple"]

for tree in parser.parse(sentence):
    print(tree)
```
